```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib.backends.backend_qt5agg import FigureCanvas
import numpy as np
```

```python
x = np.linspace(-10, 0, 101)
y = [1]*len(x)

fig = plt.figure(dpi=120)
ax = fig.subplots()
line, = ax.plot(x, y)
plt.show()
```

```python

new_y = 2

y.pop(0)
y.append(new_y)

fig = plt.figure(dpi=120)
ax = fig.subplots()
line, = ax.plot(x, y)
line.set_data(x, y)
plt.show()
```
